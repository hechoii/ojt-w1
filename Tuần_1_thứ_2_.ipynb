Viết class LinearLayer thuần Python

Viết một class LinearLayer không dùng PyTorch, chỉ dùng NumPy. Áp dụng đầy đủ
Type Hinting.
*   Constructor nhận in_features: int, out_features: int
*   Phương thức forward(x: np.ndarray) -> np.ndarray thực hiện xW + b
*   Phương thức __repr__() trả về string mô tả layer

```python
class LinearLayer:
    def __init__(self, int_feature: int, out_feature: int) -> None:
        self.W = np.random.rand(int_feature, out_feature)*0.01
        self.b = np.random.rand(out_feature)*0.01
        self.int_feature = int_feature
        self.out_feature = out_feature

    def forward(self, input: np.ndarray) -> np.ndarray:
        return input @ self.W + self.b

    def __repr__(self) -> str:
        return f"LinearLayer(input={self.int_feature}, out={self.out_feature})"
```

```python
layer = LinearLayer(4,2)
x= np.random.rand(8,4)
out =layer.forward(x)
assert out.shape == (8,2)
print("Pass. Output Shape:", out.shape)
print(layer)
```

Kế thừa để tạo MLPClassifier

Kế thừa từ LinearLayer, thêm:
*   Class Activation với method relu(x) và sigmoid(x)
*   Class MLPClassifier(Activation) chứa 2 LinearLayer
*   Phương thức predict(x: np.ndarray) -> np.ndarray chạy forward toàn bộ

```python
class Activation:
    def __init__(self) -> None:
        pass
    def relu(self, x):
        return np.maximum(0, x)
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
```

```python
class MDPClassifier():
    def __init__(self) -> None:
        layer = LinearLayer(4,2)
        self.layers = [layer]
    def forward(self, x) -> np.ndarray:
        k = self.layers[0].forward(x)
        for layer in self.layers[1:]:
            k = layer.forward(k)
        return k
    def predict(self, x) -> np.ndarray:
        output = self.forward(x)
        return np.argmax(output, axis=1)
```

```python
layer = LinearLayer(4,2)
x= np.random.rand(8,4)
classifier = MDPClassifier()
x = classifier.predict(x)
print(x)
```

Viết SimpleMLP kế thừa nn.Module

Viết một mạng MLP 3 tầng dùng PyTorch với đầy đủ Type Hinting.

*   Input: 784 chiều (MNIST flatten), Hidden: 256, Output: 10 class
*   Dùng nn.Linear, nn.ReLU, nn.Dropout(p=0.3)
*   Override forward(self, x: torch.Tensor) -> torch.Tensor

```python
class SimpleMLP(nn.Module):
    def __init__(
        self,
        input_dim: int = 784,
        hidden_dim: int = 256,
        num_classes: int = 10,
        dropout: float = 0.3,
    ) -> None:
        super().__init__()
        # TODO: định nghĩa self.net = nn.Sequential(...)
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, num_classes)
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # TODO: flatten x rồi đưa qua self.net
        x = torch.flatten(x, start_dim=1)
        return self.net(x)
```

```python
model = SimpleMLP()
dummy = torch.randn(32, 1, 28, 28)  # batch=32, MNIST shape
out = model(dummy)
assert out.shape == (32, 10), f"Shape sai: {out.shape}"
print("✅ Pass! Output shape:", out.shape)
print(model)
```

Viết hàm count_params(model: nn.Module) -> dict trả về:

*   total: tổng số params
*   trainable: số params có requires_grad=True

In ra: "SimpleMLP: 235,146 params (235,146 trainable)"

```python
def count_params(model: nn.Module) -> dict:
    total = 0
    trainable = 0
    for param in model.parameters():
        total += param.numel()
        if param.requires_grad:
            trainable += param.numel()
    return f"SimpleML: {total} params ({trainable} trainable)"
```

```python
count_params(model)
```
